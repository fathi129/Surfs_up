```python
from sqlalchemy import create_engine,inspect
import pandas as pd

# Path to sqlite, THIS MAY NOT MATCH YOUR PATH
database_path = "/Users/fathi/Downloads/Census_Data.sqlite"

# Create an engine that can talk to the database
engine = create_engine(f"sqlite:///{database_path}")
conn = engine.connect()
```

```python
inspector_gadget = inspect(engine)
tables = inspector_gadget.get_table_names()
for table in tables:
    print(table)

    columns = inspector_gadget.get_columns(table)
    for col in columns:
        print(col['name'],col['type'])
```

```python
census_data_df = pd.read_sql("select * from Census_Data",conn)
census_data_df
```

```python
filtered_data = census_data_df[census_data_df['Population'] >= 100000].sort_values(by='Median Age')
filtered_data.head()
```
